**Notebook Documentation: weather_pulling.ipynb**

**Purpose:**

This Jupyter notebook demonstrates how to use the `weekly_weather_past()` function from the `weather_pull` module to retrieve weather data for specific geographical zones.

**Input:**

* An Excel spreadsheet containing zone information, including Easting, Northing, and Zone.
* The path to the Excel file.

**Output:**

* CSV files containing weather data for each zone specified in the Excel spreadsheet.

**Steps:**

1. **Data Import:**
    * Import the necessary libraries, including pandas, pyproj, and the `weather_pull` module.
    * Read the Excel spreadsheet using pandas, specifying the sheet names.
    * Extract the longitude, latitude, and zone information from the Excel spreadsheet.

2. **Data Transformation:**
    * Use the `Transformer` from pyproj to convert Easting and Northing coordinates to longitude and latitude.
    * Create a new column in the dataframe containing the zone information.

3. **Weather Data Retrieval:**
    * Iterate over each row in the dataframe.
    * Extract the longitude, latitude, and zone for each row.
    * Call the `weekly_weather_past()` function with the extracted coordinates and zone.
    * Save the weather data to a CSV file with a filename based on the zone.

4. **Data Visualization:**
    * Read the CSV file containing the weather data.
    * Print the data to display the retrieved weather information.

````python
import sys
from pathlib import Path

import pandas as pd
from pyproj import Transformer

sys.path.append("../src")
from weather_pull import weekly_weather_past
````

````python
# Import provided data from multi-sheet xlsx files
folderpath = (
    Path("..")
    / "data"
    / "The Water Sector Meets Mathematics_ V-KEMS Virtual Study Group-20260323T171309Z-3-001"
    / "Challenge 1"
)
filepath = folderpath / "VKEMS challenge 1 data.xlsx"
with open(filepath, "r") as file:
    df = pd.read_excel(filepath, sheet_name=None)

# Show imported Excel sheet names
print(df.keys())
````

````python
transformer = Transformer.from_crs("EPSG:27700", "EPSG:4326", always_xy=True)

df["Zones"]["longitude"], df["Zones"]["latitude"] = transformer.transform(
    df["Zones"]["Easting"], df["Zones"]["Northing"]
)
df["Zones"] = df["Zones"].set_index("Zone")
df["Zones"].head()
````

````python
# From each row, extract longitude, latitude and Zone and use it as input of a weather_pull call

for index, row in df["Zones"].iterrows():
    lat, long = row[["latitude", "longitude"]].values
    filename, df_tmp = weekly_weather_past(lat, long, index)
    if filename:
        df_tmp.to_csv(filename, index=True)
````
